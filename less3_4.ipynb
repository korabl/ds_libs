```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
%config InlineBackend.figure_format = "svg"

plt.style.use('fivethirtyeight')
```

```python
DATASET_PATH = ('creditcard.csv')
```

```python
df = pd.read_csv(DATASET_PATH)
df.head()
```

```python
df['Class'].value_counts().plot.bar(legend=True)
```

```python
df['Class'].value_counts().plot.bar(log=True)
```

```python
filter1 = df["Class"].isin(["1"]) 
filter2 = df['Class'].isin(['0'])
new = df[filter1] 
new1 = df[filter2]
```

```python
df1 = pd.DataFrame({'Class1': new['V1'], 'Class0': new1['V1']}, columns=['Class1', 'Class0'])

df1.plot.hist(alpha=0.5, log=True)
plt.xlabel('Class')
plt.title('Log распределение значений V1')
```

```python
df1.plot.hist(alpha=0.5, legend=True)
plt.xlabel('Class')
plt.title('Обычное распределение значений V1')
```

```python
fig = plt.figure(figsize=(8, 6))
ax_1 = fig.add_subplot(2, 2, 1)
ax_2 = fig.add_subplot(2, 2, 2)
ax_3 = fig.add_subplot(2, 2, 3)

ax_1.hist(df['V1'], bins=20)
ax_2.hist(new['V1'], bins=20)
ax_3.hist(new1['V1'], bins=20)


ax_1.set(title = 'Все значения V1')
ax_2.set(title = 'Мошенники')
ax_3.set(title = 'Норм')

plt.subplots_adjust(wspace=0.3, hspace=0.3)
```
